```python
df = df.replace(0,np.nan)
```

```python
df["precipitation_sum"].fillna(df["precipitation_sum"].median(),inplace=True)
```

```python
df["rain_sum"].fillna(method="ffill", inplace=True)
df["rain_sum"].fillna(method='bfill',inplace=True)
```

```python
mode_weather = df["weather_code"].mode()[0]
df["weather_code"].fillna(mode_weather,inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.to_csv("weather_2021_2022.csv")
```

```python
df1 = pd.read_csv('mumbai.csv')
```

```python
df1.head()
```

```python
df1.tail()
```

```python
df1.info()
```

```python
df1.describe()
```

```python
missing_percent = df1.isnull().sum()
print(missing_percent)
```

```python
df1 = df1.replace(0, np.nan)
```

```python
mean_cols = ["PM2.5", "PM10", "NO", "NO2", "NOx", "CO"]
df1[mean_cols] = df1[mean_cols].fillna(df1[mean_cols].mean())
```

```python
median_cols = ["NH3", "SO2", "Ozone", "Benzene", "Toluene", "Eth-Benzene", "MP-Xylene", "Xylene"]
df1[median_cols] = df1[median_cols].fillna(df1[median_cols].median())
```

```python
ffill_cols = ["RH", "WS", "WD", "AT", "BP", "Temp", "VWS"]
df1[ffill_cols] = df1[ffill_cols].fillna(method="ffill").fillna(df1[ffill_cols].mean())
```

```python
rain_cols = ["RF", "TOT-RF"]
df1[rain_cols] = df1[rain_cols].fillna(method="bfill").fillna(method="ffill")
```

```python
df1["SR"] = df1["SR"].fillna(df1["SR"].mode()[0])
```

```python
df1.to_csv("AQL_MUMBAI.csv")
```
