```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AQL_MUMBAI.csv")
```

Histogram / KDE plots of pollutants (e.g., PM2.5, PM10)

```python
sns.histplot(df["PM2.5"],kde=True,bins=50)
plt.title("Distribution of PM2.5 levels")
plt.show()
```

Trend of PM2.5 over time

```python
df['From Date'] = pd.to_datetime(df['From Date'])
plt.figure(figsize=(12,6))
sns.lineplot(x="From Date", y="PM2.5", data=df.sample(5000))
plt.title("PM2.5 Trend Over Time")
plt.show()
```

Seasonal/monthly averages

```python
df["month"] = df["From Date"].dt.month
sns.boxplot(x="month", y="PM2.5", data=df)
plt.title("Monthly Variation in PM2.5")
plt.show()
```

Correlation Heatmap (pollutants vs weather)

```python
numeric_df = df.select_dtypes(include="number")
plt.figure(figsize=(12,6))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Pollutants & Weather")
plt.show()
```

Scatterplot: Temp vs PM2.5

```python
sns.scatterplot(x="Temp", y="PM2.5", data=df.sample(5000))
plt.title("Relationship: Temperature vs PM2.5")
plt.show()
```

Wind Speed vs Pollutant Concentration

```python
sns.scatterplot(x="WS", y="PM2.5", data=df.sample(5000))
plt.title("Wind Speed vs PM2.5")
plt.show()
```

Wind Direction impact

```python
sns.boxplot(x="WD", y="PM2.5", data=df.sample(5000))
plt.title("PM2.5 Levels by Wind Direction")
plt.show()
```
