```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("weather_2021_2022.csv")
```

Trend over time

```python
df['date'] = pd.to_datetime(df['date'])
plt.figure(figsize=(12,6))
sns.lineplot(x="date", y="temperature_2m_max", data=df, label="Max Temp")
sns.lineplot(x="date", y="temperature_2m_min", data=df, label="Min Temp")
plt.title("Daily Temperature Trend")
plt.show()
```

Distribution of temperatures

```python
sns.histplot(df["temperature_2m_max"], kde=True)
plt.title("Max Temperature Distribution")
plt.show()
```

Monthly rainfall trend

```python
df['month']=df['date'].dt.month
sns.barplot(data=df,x="month",y="rain_sum",estimator=sum)
plt.title("Monthly rainfall in(Total)")
plt.show()
```

Rain vs Temperature

```python
sns.scatterplot(data=df,x="rain_sum",y="temperature_2m_max")
plt.title("rain vs temperature")
plt.show()
```

Wind speed distribution

```python
sns.histplot(data=df, x="wind_speed_10m_max",bins=30,kde=True)
plt.title("wind speed distripution")
plt.show()
```

Wind vs Precipitation

```python
sns.scatterplot(x="wind_speed_10m_max", y="precipitation_sum", data=df)
plt.title("Wind Speed vs Rainfall")
plt.show()
```

Heatmap of correlations

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Seasonal Boxplots (Temp by Month)

```python
sns.boxplot(x="month", y="temperature_2m_max", data=df)
plt.title("Monthly Max Temperature Distribution")
plt.show()
```
